```python
%%sql
CREATE VIEW TopCustomers AS 
SELECT c.FirstName, c.LastName, c.CustomerEmail AS Customer, Sum(s.Netsales) AS TotalDollarSpent, s.NetQuantity AS QuantityPurchased
FROM Contact c, Sales s
WHERE C.CustomerEmail = s.CustomerEmail
GROUP BY c.CustomerEmail;
```

```python
%%sql
SELECT * FROM TopCustomers
WHERE TotalDollarSpent > 300;
```

```python
#Create View to see who is the best salesmen
```

```python
%%sql
CREATE VIEW TOPSALESMEN AS
SELECT s.StaffName AS Employee, SUM(s.NetSales) AS TotalNetSales, SUM(s.NetQuantity) AS TotalQuantitySold
FROM Sales s
GROUP BY s.StaffName;
```

```python
%%sql
SELECT * FROM TOPSALESMEN
ORDER BY TotalNetSales DESC;
```

```python
%%sql
CREATE VIEW TOPSELLER AS
SELECT s.ItemID, i.DCS , i.Style, i.Category, i.ColorDes, SUM(s.NetQuantity) AS TotalQuantitySold
FROM Sales s, Inventory i
WHERE s.ItemID = i.ItemID
GROUP BY s.ItemID;
```
